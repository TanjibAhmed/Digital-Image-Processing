```python
############ Video of Zooming the image

import cv2
import numpy as np

# Set image file and output video file
image_file = r'sol.webp'
output_video_file = r'C:\Users\Tanjib Ahmed\Desktop\video\output_video.avi'

class transformation:
    def __init__(self, loc, output_video_file, method):
        # Load the image
        imag = cv2.imread(loc)
        self.imag = imag
        self.output_video_file = output_video_file
        if imag is None:
            print("Error: Unable to load the image.")
            exit()

        # Define video properties
        self.frame_width = imag.shape[1]
        self.frame_height = imag.shape[0]
        self.fps = 30  # Frames per second
        self.duration = 10  # Duration of the video in seconds
        self.image = []

        # Calculate the center of the image
        self.center_x = self.frame_width // 2
        self.center_y = self.frame_height // 2
        
        if method == 'zoom':
            self.image = self.zoom()
        elif method == 'rotation':
            self.image = self.rotation(imag)
        elif method == 'shear':
            self.image = self.shear()
        self.video()
        print(f"Video '{output_video_file}' created successfully.")
        

    def zoom(self):
        zoom_steps = 100  # Number of zoom steps
        for step in range(zoom_steps):
            zoom_factor = 1 + step / zoom_steps  # Adjust the zoom factor

            # Adjust the center of the zoom operation to the middle of the image
            M = np.float32([[zoom_factor, 0, (1 - zoom_factor) * self.center_x],  #instead of (0,0) position i am starting from the middle point
                            [0, zoom_factor, (1 - zoom_factor) * self.center_y],
                            [0, 0, 1]])

            dst = cv2.warpPerspective(self.imag, M, (self.frame_width, self.frame_height))
            self.image.extend([dst] * (self.fps // self.duration))  # Add frames for the duration of each zoom step. to create a video effect where a single frame is displayed for a certain duration.
            #this extend is used so that the a frame remains same for a certain period of time.
            #extend is used to add multiple data point
        return self.image
    
    def rotation(self):
        for i in range(0, 360):   
            M = cv2.getRotationMatrix2D(((self.frame_width-1)/2.0,(self.frame_height-1)/2.0),i,1)
            #print(M)
            a = [[0, 0 , 1]]
            T = np.append(M, a, axis=0)
            dst = cv2.warpPerspective(self.imag,T,(int(self.frame_width*2),int(self.frame_height*2))) 
            self.image.append(dst)
        return self.image
    
    def shear(self):
        sheared_steps = 100
        for step in range(sheared_steps):
            s_f =   step / sheared_steps  # Adjust the zoom factor

            # Adjust the center of the zoom operation to the middle of the image
            M = np.float32([[1, s_f, 0],  
                            [0, 1, 0],
                            [0, 0, 1]])

            dst = cv2.warpPerspective(self.imag, M, (int(self.frame_width*1.5),int(self.frame_height*1.5)))
            self.image.extend([dst] * (self.fps // self.duration))  # Add frames for the duration of each zoom step. to create a video effect where a single frame is displayed for a certain duration.
            #this extend is used so that the a frame remains same for a certain period of time.
        return self.image

    def video(self):
        # Create VideoWriter object
        fourcc = cv2.VideoWriter_fourcc(*'XVID')  #specifying video compression algorithm. gives some int values
        out = cv2.VideoWriter(self.output_video_file, fourcc, self.fps, (self.frame_width, self.frame_height))
        # Write frames to the video
        for img in self.image:
            out.write(img)

        # Release the VideoWriter object
        return out.release()

# Set image file and output video file
image_file = r'sol.webp'
output_video_file = r'C:\Users\Tanjib Ahmed\Desktop\video\output_video5.avi'

transformation(loc = image_file, output_video_file = output_video_file, method = 'shear')
```
